```python
# Diketahui 2 titik
def garis_dua_titik(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    c = y1 - m * x1
    return m, c

x1 = float(input("Masukkan koordinat x untuk titik pertama: "))
y1 = float(input("Masukkan koordinat y untuk titik pertama: "))
x2 = float(input("Masukkan koordinat x untuk titik kedua: "))
y2 = float(input("Masukkan koordinat y untuk titik kedua: "))

m, c = garis_dua_titik(x1, y1, x2, y2)

print(f"Persamaan garis: y = {m}x + {c}")
```

```python
# Diketahui 1 gradien dan 1 titik yang dilewati
def garis_gradien_dan_titik(m, x, y):
    c = y - m * x
    return m, c

m = float(input("Masukkan gradien: "))
x = float(input("Masukkan koordinat x untuk titik: "))
y = float(input("Masukkan koordinat y untuk titik: "))

m, c = garis_gradien_dan_titik(m, x, y)

print(f"Persamaan garis: y = {m}x + {c}")
```

```python
# Diketahui persamaan ax + by = c
def garis_persamaan(a, b, c):
    m = -a / b
    c = c / b
    return m, c

a = float(input("Masukkan nilai a: "))
b = float(input("Masukkan nilai b: "))
c = float(input("Masukkan nilai c: "))

m, c = garis_persamaan(a, b, c)

print(f"Persamaan garis: y = {m}x + {c}")
```
